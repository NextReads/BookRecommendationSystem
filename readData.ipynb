```python
import gzip
import json
import os
import numpy as np
import pandas as pd
```

```python
def load_data(file_name):
    data = []
    with gzip.open(file_name) as f:
        for l in f:
            d = json.loads(l.strip())
            new_d= {k: d[k] for k in ('user_id', 'book_id', 'review_id', 'rating', 'review_text')}
            data.append(new_d)
    return data

def load_data2(file_name):
    data = []
    with gzip.open(file_name) as f:
        for l in f:
            d = json.loads(l.strip())
            new_d ={k: d[k] for k in ('isbn', 'text_reviews_count', 'average_rating', 'description', 'link', 'url', 'image_url' ,'book_id','ratings_count','title')}
            data.append(new_d)
    return data
```
